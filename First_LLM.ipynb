```python
from langchain_groq import ChatGroq
from google.colab import userdata
llm_groq = ChatGroq(model_name="llama-3.3-70b-versatile", api_key=userdata.get("GROQ_API_KEY"), temperature=0.7)
llm_groq2 = ChatGroq(model_name="openai/gpt-oss-120b", api_key=userdata.get("GROQ_API_KEY"), temperature=0.7)
```

```python
ai_mesg = llm_groq.invoke("When was Michael Jackson born")
print(ai_mesg.content)
```

```python
ai_mesg = llm_groq2.invoke("create a marketing strategy for my startup on machine learning")
print(ai_mesg.content)
```

```python
ai_mesg = llm_groq2.invoke("Write me a random Dotnet C# code for invoice processing in insurance/healthcare")
print(ai_mesg.content)
```
