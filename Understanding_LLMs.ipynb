```python
import random

def get_simple_sentences():
    """Return predictable sentences for beginners"""
    return [
        "The cat sat on the mat and fell asleep.",
        "I like to eat pizza for dinner on Friday nights.",
        "The sun rises in the east every single morning.",
        "She went to the store to buy milk and bread.",
        "My favorite color is blue because it reminds me of the ocean.",
        "The dog barked loudly at the mailman walking by.",
        "We drove to school in our red car this morning.",
        "The book was so interesting that I read it all night.",
        "It started raining heavily so we ran inside the house.",
        "My mom makes the best chocolate chip cookies in town."
    ]

def play_continuous_game():
    """Play continuous token prediction on the same sentence"""
    sentences = get_simple_sentences()

    print("🤖 Welcome to Continuous Token Prediction!")
    print("Predict one word at a time as the sentence builds up.")
    print("This is exactly how LLMs generate text!\n")

    total_score = 0
    total_predictions = 0

    # Play 3 sentences
    for round_num in range(1, 4):
        sentence = random.choice(sentences)
        words = sentence.split()

        print(f"--- Sentence {round_num}/3 ---")
        print("Starting with:", f'"{words[0]} {words[1]} {words[2]}"')

        # Start predicting from the 4th word
        revealed = words[:3]  # Show first 3 words
        round_score = 0
        round_predictions = 0

        for i in range(3, len(words)):
            target_word = words[i]

            print(f'\nContext: "{" ".join(revealed)}"')
            print(f"Predict next word (hint: {target_word[0]}...):")

            guess = input("> ").strip().lower()
            actual = target_word.lower().rstrip('.,!?')

            if guess == actual:
                print(f"✅ Correct! Next word: '{target_word}'")
                round_score += 1
            else:
                print(f"❌ The word was: '{target_word}'")

            # Add the word to revealed context
            revealed.append(target_word)
            round_predictions += 1

            # Show growing sentence
            print(f'Sentence so far: "{" ".join(revealed)}"')

        print(f"\n✨ Complete sentence: \"{sentence}\"")
        print(f"Round score: {round_score}/{round_predictions}")

        total_score += round_score
        total_predictions += round_predictions
        print()

    # Final results
    accuracy = (total_score / total_predictions) * 100 if total_predictions > 0 else 0
    print(f"🎯 Final Score: {total_score}/{total_predictions} ({accuracy:.0f}%)")

    if accuracy >= 60:
        print("🏆 Amazing! You're predicting like a pro LLM!")
    elif accuracy >= 40:
        print("👍 Great job! You understand language patterns!")
    elif accuracy >= 20:
        print("📈 Good start! Keep practicing token prediction!")
    else:
        print("🤔 Language prediction is tricky - even for AI!")

    print("\n💡 This is exactly how LLMs work:")
    print("   • They see the context so far")
    print("   • They predict the most likely next token")
    print("   • They add it and predict the next one")
    print("   • This continues until the text is complete!")
```

## 2. Testing token prediction
Autoregressive generation

```python
from transformers import AutoModelForCausalLM
model_name = "HuggingFaceTB/SmolLM2-135M-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to("cpu")
```

```python
import torch
import torch.nn.functional as F
def predict_next_token(text, num_tokens=5, temperature=0):
    # Convert text to model format
    tokens = tokenizer.encode(text, return_tensors="pt")

    # Get model prediction
    output = model(tokens)

    # Focus on the last position (next token)
    next_token_scores = output.logits[0, -1, :]

    #Apply temperature (higher = more random)
    if temperature > 0:
        next_token_scores = next_token_scores / temperature

    # Convert scores to probabilities (0-1)
    probabilities = F.softmax(next_token_scores, dim=-1)

    # Get top predictions
    top_probs, top_ids = torch.topk(probabilities, num_tokens)

    # Show results
    print(f"After '{text}', the model predicts:")
    print("-" * 80)
    print("Top tokens possible: ",top_ids)
    for i, (prob, token_id) in enumerate(zip(top_probs, top_ids)):
        token_text = tokenizer.decode(token_id)
        percentage = prob * 100
        print(f"{i+1}. '{token_text}',  - {percentage:.1f}%, id: {token_id}")

    # Sample the next token based on the probability distribution
    next_token_id = torch.multinomial(probabilities, num_samples=1).item()
    next_token_text = tokenizer.decode(next_token_id)
    next_token_prob = probabilities[next_token_id].item() * 100

    print(f"FINAL PREDICTION: '{next_token_text}' ({next_token_prob:.1f}%)")
    return next_token_text
```

```python
def generate_text(text, max_length=100, top_k=5,temperature=1):
    # Keep track of original input
    original_text = text
    generated_tokens = 0

    # Generate tokens one by one (autoregressive)
    for i in range(1, max_length + 1):
        # Predict next token using our earlier function
        new_token = predict_next_token(text, top_k, temperature)

        text = text + new_token
        generated_tokens += 1

        # Show progress
        print(f"Token {i}: '{new_token}'")
        print(f"Text so far: '{text}'")
        print("-" * 40)

        # Stop if we get an end marker
        if new_token == "<|endoftext|>":
            break

    print(f"\nFINAL RESULT:")
    print(f"Original prompt: '{original_text}'")
    print(f"Generated {generated_tokens} new tokens")
    print(f"Complete text: '{text}'")

    return text
```
